<h5 style="color:brown">Data preprocessing</h5>

```python
df.drop("Address",axis=1,inplace=True)
```

<h3 style="color:orange">Data validation(splitting)</h3>

```python
x=df.drop("Price",axis=1)
```

```python
x.head()
```

```python
y=df["Price"]
```

```python
y.head()
```

<h3 style="color:red">Model Building</h3>

```python
#Random_state is the seed used by the random number generator
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,random_state=24)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred
```

<h3 style="color:brown">Model Evaluation</h3>

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
mse=mean_squared_error(y_test,y_pred)
```

```python
mse
```

```python
rmse=sqrt(mean_squared_error(y_test,y_pred))
```

```python
rmse
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```

.In this evaluation we observed r2_score is good and errors are low hence the model is good fit
